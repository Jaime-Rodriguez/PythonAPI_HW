### Analysis
* As expected, the weather becomes significantly warmer as one approaches the equator (0 Deg. Latitude). More interestingly, however, is the fact that the southern hemisphere tends to be warmer this time of year than the northern hemisphere. This may be due to the tilt of the earth.
* There is no strong relationship between latitude and cloudiness. However, it is interesting to see that a strong band of cities sits at 0, 80, and 100% cloudiness.
* There is no strong relationship between latitude and wind speed. However, in northern hemispheres there is a flurry of cities with over 20 mph of wind.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import openweathermapy.core as owm
import urllib, json
from datetime import datetime
import os
from citipy import citipy


# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy


# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
city_df = pd.DataFrame(cities,columns=['Cities'])
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
lon = []
lat = []
country = []
hum = []
maxtemp = []
clouds = []
wind = []
new_city = []
dt = []
for index, row in city_df.iterrows():


    try:    
        # get restaurant type from df
        city = row['Cities']
        settings = {"units": "metric", "appid": api_key}
        json_df = owm.get_current(city, **settings)
        
        # Append corresponding values to their respective lists
        new_city.append(row['Cities'])
        lon.append(json_df['coord']['lon'])
        lat.append(json_df['coord']['lat'])
        city = citipy.nearest_city(json_df['coord']['lat'],json_df['coord']['lon'])
        country.append(city.country_code)
        hum.append(json_df['main']['humidity'])
        maxtemp.append(json_df['main']['temp_max'])
        clouds.append(json_df['clouds']['all'])
        wind.append(json_df['wind']['speed'])
        dt.append(json_df['dt'])
        time.sleep(1)     

    # For any errors (specifically the 404 Error) move on to the next record    
    except:
        next
    print([row])
         
         
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Create a dataframe with all the attributes

city_dict = {'City':new_city,'Cloudiness':clouds,'Country':country,'Date':dt,'Humidity':hum,'Lat':lat,'Lng':lon,'Max Temp':maxtemp,'Wind Speed':wind}
city_df = pd.DataFrame(city_dict)

# Clean up the columns a bit 

city_df['City'] = city_df['City'].str.title()
city_df['Country'] = city_df['Country'].str.upper()
city_df
```

###### Output the data to a csv 

```python
# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Check to see if the output folder has been created. If not, use mkdir to create the folder
outname = 'cities.csv'
outdir = './output'
if not os.path.exists(outdir):
    os.mkdir(outdir)

fullname = os.path.join(outdir, outname)    

city_df.to_csv(fullname)
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

###### Define the font weights I'll be using in my plots 

```python
font = {'weight' : 'bold',
        'size'   : 22}
font2 = {'weight' : 'bold',
        'size'   : 12}
font3 = {'weight' : 'normal',
        'size'   : 9}
```

###### Define today's run date for our plots 

```python
todayDate = '('+datetime.today().strftime('%m-%d-%Y')+')'
```

#### Latitude vs. Temperature Plot

```python
plt.scatter(lat, maxtemp,linewidths=1, edgecolors='black',alpha=.9)
plt.title('Latitude vs. Temperature Plot '+todayDate,**font2)
plt.xlabel('Latitude',**font3)
plt.ylabel('Max Temperature (F)',**font3)
plt.grid()
```

#### Latitude vs. Humidity Plot

```python
plt.scatter(lat, hum,linewidths=1, edgecolors='black',alpha=.9)
plt.title('Latitude vs. Humidity Plot '+todayDate,**font2)
plt.xlabel('Latitude',**font3)
plt.ylabel('Humidity (%)',**font3)
plt.grid()
```

#### Latitude vs. Cloudiness Plot

```python
plt.scatter(lat, clouds,linewidths=1, edgecolors='black',alpha=.9)
plt.title('Latitude vs. Cloudiness Plot '+todayDate,**font2)
plt.xlabel('Latitude',**font3)
plt.ylabel('Cloudiness (%)',**font3)
plt.grid()
```

#### Latitude vs. Wind Speed Plot

```python
plt.scatter(lat, wind,linewidths=1, edgecolors='black',alpha=.9)
plt.title('Latitude vs. Wind Speed '+todayDate,**font2)
plt.xlabel('Latitude',**font3)
plt.ylabel('Wind Speed (MPH)',**font3)
plt.grid()
```

```python
fig, ax = plt.subplots(2,2, figsize = (15,10))

fig.suptitle('Latitude and Weather '+todayDate,**font)

ax[0,0].scatter(lat, maxtemp,linewidths=1, edgecolors='black',alpha=.9)
ax[0,0].set_title('Latitude vs. Temperature Plot',**font2)
ax[0,0].set_xlabel('Latitude',**font3)
ax[0,0].set_ylabel('Max Temperature (F)',**font3)
ax[0,0].grid()

ax[1,0].scatter(lat, hum,linewidths=1, edgecolors='black',alpha=.9)
ax[1,0].set_title('Latitude vs. Humidity Plot',**font2)
ax[1,0].set_xlabel('Latitude',**font3)
ax[1,0].set_ylabel('Humidity (%)',**font3)
ax[1,0].grid()

ax[0,1].scatter(lat, clouds,linewidths=1, edgecolors='black',alpha=.9)
ax[0,1].set_title('Latitude vs. Cloudiness Plot',**font2)
ax[0,1].set_xlabel('Latitude',**font3)
ax[0,1].set_ylabel('Cloudiness (%)',**font3)
ax[0,1].grid()

ax[1,1].scatter(lat, wind,linewidths=1, edgecolors='black',alpha=.9)
ax[1,1].set_title('Latitude vs. Wind Speed',**font2)
ax[1,1].set_xlabel('Latitude',**font3)
ax[1,1].set_ylabel('Wind Speed (MPH)',**font3)
ax[1,1].grid()

fig.tight_layout(rect=[0,0.03,1,0.95])
```
